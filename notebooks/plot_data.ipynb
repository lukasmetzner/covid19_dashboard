```python
import pandas as pd
import plotly.express as px
import scipy.signal as sig

df_cases = pd.read_csv("../cases_transformed.csv")
df_deaths = pd.read_csv("../deaths_transformed.csv")
df_recovered = pd.read_csv("../recovered_transformed.csv")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=3, cols=2)
fig.append_trace(go.Line(x=df_cases.date.values, y=df_cases.new_cases.values, name="Cases"), row=1, col=1)
fig.append_trace(go.Line(x=df_cases.date.values, y=sig.savgol_filter(df_cases.new_cases, 33, 3), name="Cases Smoothed"), row=1, col=2)

fig.append_trace(go.Line(x=df_deaths.date.values, y=df_deaths.new_deaths.values, name="Deaths"), row=2, col=1)
fig.append_trace(go.Line(x=df_deaths.date.values, y=sig.savgol_filter(df_deaths.new_deaths, 33, 3), name="Deaths Smoothed"), row=2, col=2)

fig.append_trace(go.Line(x=df_recovered.date.values, y=df_recovered.new_recovered.values, name="Recovered Smoothed"), row=3, col=1)
fig.append_trace(go.Line(x=df_recovered.date.values, y=sig.savgol_filter(df_recovered.new_recovered, 33, 3), name="Recovered"), row=3, col=2)
fig.update_layout(height=600, width=800, title_text="Smoothed and raw corona data")
fig.show()
```
